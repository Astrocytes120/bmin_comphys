## 과제 1 광전효과

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이캠퍼스에서 다음의 데이터를 다운받아서 fitting하세요.
# Photon frequencies and stopping voltages from Millikan's photoelectric experiment
# 이 데이터는 밀리컨이 수행한 광전효과 실험 결과이다. 
# xdata는 진동수 [Hz], ydata는 전압 [V]이다.
data=np.loadtxt("./data/millikan.txt",float)
xdata=[i[0] for i in data] 
ydata=[i[1] for i in data]
plt.xlabel('frequency (Hz)')
plt.ylabel('Voltage (V)')
plt.plot(xdata,ydata,'-o');
```
