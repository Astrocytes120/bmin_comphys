## 과제 1 광전효과

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이캠퍼스에서 다음의 데이터를 다운받아서 fitting하세요.
# Photon frequencies and stopping voltages from Millikan's photoelectric experiment
# 이 데이터는 밀리컨이 수행한 광전효과 실험 결과이다. 
# xdata는 진동수 [Hz], ydata는 전압 [V]이다.
data=np.loadtxt("./data/millikan.txt",float)
xdata=np.array([i[0] for i in data])
ydata=np.array([i[1] for i in data])
plt.xlabel('frequency (Hz)')
plt.ylabel('Voltage (V)')
plt.plot(xdata,ydata,'o');
```

```python
import scipy.optimize

def func(x,a,b):
    return a*x+b
popt, pcov = scipy.optimize.curve_fit(func,xdata,ydata)
popt[0]
```

```python
plt.xlabel('frequency (Hz)')
plt.ylabel('Voltage (V)')
plt.plot(xdata,ydata,'o');
plt.plot(xdata,func(xdata,*popt),'-');
```

## 과제 2 라이프니츠 파이 공식

\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots
\end{align}
이 식은 라이프니츠의 파이 공식이다.  
1. 우변의 합을 구하는 함수를 만들어라.  
2. 이 식을 이용하여 아래처럼 그림을 그려라.

```python
import numpy as np

def lei(n):
    i, s=1, 0
    while(i<=n):
        if(i%4==1):
            s=s+(1/i)
        else:
            s=s-(1/i)
        i=i+2
    return s

x=np.arange(1,100,2)
y=np.array([lei(i) for i in x])
plt.xlabel('n')
plt.plot(x,y,label='Leibniz formula for $\pi$')
plt.plot([1,100],[np.pi/4,np.pi/4],label='$\pi/4$') 
plt.legend();
plt.savefig('./leibniz.pdf')
```
