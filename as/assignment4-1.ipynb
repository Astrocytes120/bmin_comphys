# Gamma function
* $\Gamma(z) = \int_0^{\infty} e^{-x} x^{z-1} dx = (z-1)!$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
def gam(x,n):
    return np.exp(-x)*x**(n-1)
integrate.quad(gam, 0., np.inf, args=4)
```

```python
import math
xdata = np.arange(1,11.2,0.1)
ydata = [integrate.quad(gam,0,np.inf,args=i)[0] for i in xdata]

plt.plot(xdata,ydata,'-',label='gamma')
plt.plot(np.arange(1,12,1),[math.factorial(i-1) for i in np.arange(1,12,1)],'o',label='factorial')
plt.legend();
```

# Riemann-zeta function and Polylogarithm
* $\zeta(s)=\sum_{n=1}^{\infty} \frac{1}{n^s}$
* $\zeta(s)=\frac{1}{\Gamma(s)} \int_0^{\infty} \frac{x^{s-1}}{e^x-1} dx $

### Check $\int_0^{\infty} \frac{x^{3}}{e^x-1} dx = \frac{\pi^4}{15}$
* ref: https://en.wikipedia.org/wiki/Polylogarithm

```python
def pl(x):
    return x**3/(np.exp(x)-1)
integrate.quad(pl, 0, 10), np.pi**4/15
```

# Harmonic Oscillators  
* $\ddot{y} + \omega^2 y =0$ where $\omega=\sqrt{\frac{k}{m}}.$
* $y_0=1$, $v_0=0$, $\omega=1$, $a=-y$.

```python
dt=0.001
def dy(y,v):
    a=-y
    return [v*dt+0.5*a*dt**2,-y*dt]

y,v=1,0
tdata=np.arange(0,10,dt)
ydata=[]
for i in tdata:
    ydata.append([y,v])
    y,v = y+dy(y,v)[0],v+dy(y,v)[1]
```

```python
plt.plot(tdata,[i[0] for i in ydata],label='y');
plt.plot(tdata,[i[1] for i in ydata],label='v');
plt.legend();
```
