# 4. Harmonic Oscillators  
* $\ddot{y} + \omega^2 y =0$ where $\omega=\sqrt{\frac{k}{m}}.$
* $y_0=1$, $v_0=0$, $\omega=1$, $a=-y$.

* 위와 같이 주어진 조화진동자의 미분방정식을 미분 개념을 이용하여 $y(t)$, $v(t)$ 그래프를 그려라.
* 삼각함수와 exp함수는 사용하지 말 것.
* 순간 가속도가 $a$로 주어져 있을 때 속도와 변위의 변화는 다음처럼 주어진다.   
$\Delta v = a \Delta t,$  
$\Delta x = v \Delta t +\frac{1}{2} a (\Delta t)^2.$
