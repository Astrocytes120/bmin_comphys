# Dimensionality Reduction
* unsupervised learning
* high dimension의 데이터에서 중요한 features를 찾는 방법

## Pricipal Component Analysis
1. Covariance Matrix
2. Eigenvectors of covariance matrix
3. Rearrange the data with principal components

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris() # 분꽃
iris.target_names # 3종류의 분꽃 분류하기
iris.feature_names # features
```

```python
Xall=iris["data"] 
y=iris["target"] # setosa=0, versicolor=1, virginica=2
X=[Xall[:,2],Xall[:,3]]
plt.scatter(X[0],X[1],c=y)
plt.plot([1,7],[0,2.5],'k-')
```

### sklearn.decomposition.PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1)
```

```python
pca.fit(np.transpose(X))
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
plt.scatter(X[0],X[1],c=y)
plt.plot([0,6*pca.components_[0,0]],[0,6*pca.components_[0,1]],'k-')
```
